```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("chicago_traffic_crashes_clean.csv")
df.head()
df.info()
```

```python
df.duplicated().sum()
df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df_sorted = df.sort_values(by="LIGHTING_CONDITION", ascending = False)
df_sorted.info()
```

```python
df.describe()
```

```python
top_causes = df['PRIM_CONTRIBUTORY_CAUSE'].value_counts().nlargest(15)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_causes.index, y=top_causes.values, palette='viridis')
plt.title('Top Crash Causes')
plt.xlabel('Primary Cause')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
