```python
top_causes = df['PRIM_CONTRIBUTORY_CAUSE'].value_counts().nlargest(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_causes.index, y=top_causes.values, palette='viridis')
plt.title('Top Crash Causes')
plt.xlabel('Primary Cause')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Ensure 'CRASH_DATE' is datetime (idempotent operation)
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])

# Get the top 11 primary contributory causes to ensure we can exclude 'UNABLE TO DETERMINE' and still get 10
top_causes_all = df['PRIM_CONTRIBUTORY_CAUSE'].value_counts().nlargest(11).index.tolist()

# Exclude 'UNABLE TO DETERMINE' and take the first 10 causes
if 'UNABLE TO DETERMINE' in top_causes_all:
    top_causes_all.remove('UNABLE TO DETERMINE')
top_10_causes_filtered = top_causes_all[:10]

# Filter the DataFrame to include only these top 10 causes
df_top_causes_filtered = df[df['PRIM_CONTRIBUTORY_CAUSE'].isin(top_10_causes_filtered)].copy()

# Group by date and primary contributory cause, then count crashes
daily_crashes_by_cause_filtered = df_top_causes_filtered.groupby(['CRASH_DATE', 'PRIM_CONTRIBUTORY_CAUSE']).size().reset_index(name='CRASH_COUNT')
```

```python
# Create an interactive scatter plot (dot graph) using Plotly Express
fig = px.scatter(
    daily_crashes_by_cause_filtered,
    x='CRASH_DATE',
    y='CRASH_COUNT',
    color='PRIM_CONTRIBUTORY_CAUSE',
    title='Frequency of Top 10 Crash Causes Over Time (Excluding "Unable to Determine")',
    labels={'CRASH_DATE': 'Date', 'CRASH_COUNT': 'Number of Crashes', 'PRIM_CONTRIBUTORY_CAUSE': 'Primary Cause'},
    hover_data={'CRASH_DATE': '|%Y-%m-%d %H:%M:%S', 'CRASH_COUNT': True, 'PRIM_CONTRIBUTORY_CAUSE': True}
)

fig.update_layout(xaxis_title='Date', yaxis_title='Number of Crashes')
fig.show()
```
